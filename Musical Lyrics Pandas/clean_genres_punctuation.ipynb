```python
import pandas as pd
```

```python
lyrics_data = "billboard_lyrics.csv"
```

```python
lyrics_df = pd.read_csv(lyrics_data)
```

```python
lyrics_df = lyrics_df.dropna()
lyrics_df
```

```python
test = lyrics_df.sort_values(by='Peak Position', ascending=True)
```

```python
drop_dupes = test.drop_duplicates(subset="SongID")
drop_dupes.head()
```

```python
replace_test = drop_dupes['Lyrics'].str.replace('\n',' ')
```

```python
drop_dupes["Clean Lyrics"] = replace_test
drop_dupes.head()
```

```python
clean_lyrics = drop_dupes["Clean Lyrics"].replace(r'\\','', regex=True)
```

```python
clean_lyrics.iloc[0]
```

```python
drop_dupes["Clean Lyrics"] = clean_lyrics
drop_dupes.head()
```

```python
clean_lyrics_df = drop_dupes.drop(columns=["Unnamed: 0","Lyrics"])
clean_lyrics_df.head()
```

```python
# removing songs by PSY -- contains korean characters/words
clean_lyrics_df = clean_lyrics_df[~lyrics_df["Performer"].str.contains("PSY")]
clean_lyrics_df.head()
```

```python
# removing punctuation from clean lyrics column
import re
p = re.compile(r'[^\w\s]+')
clean_lyrics_df['Clean Lyrics'] = [p.sub('', x) for x in clean_lyrics_df['Clean Lyrics'].tolist()]
```

```python
# test["Clean Lyrics"].iloc[0]
clean_lyrics_df.head()
```

```python
spotify_data = "spotify_data.csv"
spotify_df = pd.read_csv(spotify_data)
spotify_df.head()
```

```python
merged_df = pd.merge(clean_lyrics_df, spotify_df, how='inner', on='SongID')
pd.set_option('display.max_columns', None)
merged_df
```

```python
merged_df.info()
```

```python
merged_df["spotify_genre"]
```

```python
# split genres up into each subcategory and count
genre_list=[]
genre_count={}
for each_song in merged_df["spotify_genre"]: 
    try: 
        genres=each_song.split(", ")
        for each_genre in genres: 
            if each_genre in genre_count:
                genre_count[each_genre]+=1
            else: 
                genre_count[each_genre]=1
    except: 
        pass
genre_count
```

```python
for w in sorted(genre_count, key=genre_count.get, reverse=True):
    print(w, genre_count[w])
```

```python
merged_df['pop']=merged_df["spotify_genre"].str.contains('pop')
merged_df['dance']=merged_df["spotify_genre"].str.contains('dance')
merged_df['hiphop']=merged_df["spotify_genre"].str.contains('hip hop')
merged_df['rock']=merged_df["spotify_genre"].str.contains('rock')
merged_df['country']=merged_df["spotify_genre"].str.contains('country')
merged_df['metal']=merged_df["spotify_genre"].str.contains('metal')
merged_df['rnb']=merged_df["spotify_genre"].str.contains('r&b')
merged_df["rap"] = merged_df["spotify_genre"].str.contains('rap')
merged_df['pop']=merged_df["spotify_genre"].str.contains('pop')
```

```python
# merged_df["hiphop"] = merged_df["hiphop"].astype(int)
genres_df = merged_df.fillna(0)
genres_df.head()
```

```python
print(genres_df["hiphop"].value_counts())
print(genres_df["rap"].value_counts())
print(genres_df["rnb"].value_counts())
```

```python
genres_df["hiphop"] = merged_df["hiphop"].astype(float)
genres_df["hiphop"].value_counts()
```

```python
genres_df["pop"] = merged_df["pop"].astype(float)
genres_df["rock"] = merged_df["rock"].astype(float)
genres_df["country"] = merged_df["country"].astype(float)
genres_df["dance"] = merged_df["dance"].astype(float)
genres_df["rnb"] = merged_df["rnb"].astype(float)
genres_df["metal"] = merged_df["metal"].astype(float)
genres_df["rap"] = merged_df["rap"].astype(float)
```

```python
genres_df.head()
```

```python
clean_genres_df = genres_df.drop(columns=["Performer_y", "Song_y",
                                         "spotify_genre", "spotify_track_id", "spotify_track_preview_url",
                                         "spotify_track_album", "spotify_track_explicit", "spotify_track_duration_ms",
                                         "spotify_track_popularity", "danceability", "energy", "key", "loudness",
                                         "mode", "speechiness","acousticness", "instrumentalness", "liveness", "valence",
                                         "tempo", "time_signature"])
```

```python
clean_genres_df.head()
```

```python
clean_genres_df = clean_genres_df.rename(columns={"Song_x": "Song", "Performer_x": "Performer"})
```

```python
# removing rows with NaN values
clean_genres_df.dropna(axis=0, inplace=True)
print(len(clean_genres_df))
```

```python
# removing columns with non-ascii characters; need to get rid of korean characters/lyrics
clean_genres_df = clean_genres_df[~clean_genres_df['Clean Lyrics'].str.contains(r'[^\x00-\x7F]+')]
print(len(clean_genres_df))
```
