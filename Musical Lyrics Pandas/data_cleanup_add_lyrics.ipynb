```python
import pandas as pd
from PyLyrics import *
from datetime import datetime, timedelta
```

```python
billboard_csv = "songs_db.csv"
billboard_data = pd.read_csv(billboard_csv)
billboard_df = pd.DataFrame(billboard_data)
```

```python
billboard_df = billboard_df.drop(columns=["url"])
billboard_df.head()
```

```python
billboard_df = billboard_df[billboard_df["Peak Position"] <= 40]
billboard_df.loc[billboard_df['Peak Position'].idxmax()]
```

```python
# converting WeekID column from m/d/y to y/m/d format
billboard_df['WeekID']= pd.to_datetime(billboard_df['WeekID'])
billboard_df.head()
```

```python
billboard_df.info()
```

```python
billboard_df.loc[billboard_df['WeekID'].idxmax()]
```

```python
end_date_str = '2019-12-28'

# converting end_date_str from string to datetime object
end_date = datetime.strptime(end_date_str, '%Y-%m-%d')

# getting date 10 years ago from final data point
start_date = end_date - timedelta(days=365*10)

# using boolean mask to get all dates from start_date to end_date (previous 10 years)
mask = (billboard_df['WeekID'] >= start_date) & (billboard_df['WeekID'] <= end_date)

# using df.loc to reassign billboard_df to a df of the last 10 years of data
billboard_df = billboard_df.loc[mask]
```
