```
import pandas as pd
import numpy as np
```

```
# df=pd.read_csv('clean_genres.csv')
df=pd.read_csv('final_genre_lyrics.csv')
```

```
print(len(df))
# df.dropna(axis=0, inplace=True)
# print(len(df))
```

```
df.dropna(axis=0, inplace=True)
print(len(df))
```

```
df['Weeks on Chart'].dtype
```

```
# removing non ascii characters (some korean characters in Clean Lyrics)
df = df[~df['Clean Lyrics'].str.contains(r'[^\x00-\x7F]+')]
print(len(df))
```

```
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from sklearn.preprocessing import StandardScaler

# vectorizer=CountVectorizer(stop_words='english', ngram_range=(1, 5), max_features=5000)
vectorizer=TfidfVectorizer(stop_words='english', ngram_range=(1, 3), max_features=5000)

lyrics=df['Clean Lyrics']
vectorizer.fit(lyrics)
x=vectorizer.transform(lyrics)
y=df['Weeks on Chart']
```

```
vectorizer.get_feature_names()
```

```
# genres=df[['hiphop', 'pop', 'country', 'rock', 'rnb', 'metal', 'dance', 'rap']].to_numpy
```

```
# df[['hiphop', 'pop', 'country', 'rock', 'rnb', 'metal', 'dance', 'rap']].values[:, None]
genres=df[['hiphop', 'pop', 'country', 'rock', 'rnb', 'metal', 'dance', 'rap']]
```

```
from scipy.sparse import hstack
x_new=hstack((x, genres.values))
x_new
```

```
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test=train_test_split(x_new, y)
lr=LinearRegression()
lr.fit(x_train, y_train)
lr.score(x_test, y_test)
```

```
vectorizer.get_feature_names()
```

```
coef=pd.DataFrame()
coef['features']=vectorizer.get_feature_names()+list(genres.columns)
coef['weight']=lr.coef_
coef.sort_values('weight')
```

```
y_test
```

```
results=pd.DataFrame()
prediction=lr.predict(x_test)
results['pred']=prediction
results['actual']=list(y_test)
# results['actual']
results.head(25)
```

```
import pickle
pickle.dump(lr, open('lr_model.sav', 'wb'))
pickle.dump(vectorizer, open('vectorizer.sav', 'wb'))
```
